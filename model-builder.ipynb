```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('application_record.csv')
df2 = pd.read_csv('credit_record.csv')
df1.head()
```

```python
df = df1.merge(df2, on='ID')
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df[~df['OCCUPATION_TYPE'].isna()]
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

gender_counts = df['CODE_GENDER'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct=lambda pct: "{:.1f}%".format(pct))
```

```python
own_car_counts = df['FLAG_OWN_CAR'].value_counts()
plt.pie(own_car_counts, labels=own_car_counts.index, autopct=lambda pct: "{:.1f}%".format(pct))
```

```python
gender_counts = df['FLAG_OWN_REALTY'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct=lambda pct: "{:.1f}%".format(pct))
```

```python
df['STATUS'].replace({"C" : 1, "X" : 1, "0": 1,
                      "1" : 1, "2": 0, "3": 0,
                      "4" : 0, "5": 0}, inplace = True)

df['STATUS'].value_counts()
```

```python
import seaborn as sns

sns.heatmap(df[['CNT_CHILDREN', 'CNT_FAM_MEMBERS' , 'STATUS']].corr(), annot=True)
```

```python
df.drop(['CNT_CHILDREN', 'CNT_FAM_MEMBERS'], inplace=True, axis=1)
```

```python
df.head()
```

```python
df['CODE_GENDER'].replace({'M': 1, 'F': 0}, inplace=True)
df['FLAG_OWN_CAR'].replace({'Y': 1, 'N': 0}, inplace=True)
df['FLAG_OWN_REALTY'].replace({'Y': 1, 'N': 0}, inplace=True)
```

```python
df.head()
```

```python
corr = df[['CODE_GENDER', 'FLAG_OWN_CAR', 'FLAG_OWN_REALTY', 'AMT_INCOME_TOTAL', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'STATUS']].corr()
sns.heatmap(corr, annot=True)
```
